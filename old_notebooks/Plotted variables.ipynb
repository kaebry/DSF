```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dvar_hourly = pd.read_csv("../Data/data_reduced_variables.csv")
dvar_hourly
```

```python
dvar_hourly["datetime"] = pd.to_datetime(dvar_hourly['datetime'])
# set DateTime as index
dvar_hourly.set_index("datetime", inplace=True)
dvar_hourly.info()
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, dpi=120, figsize=(12,6))
for i, ax in enumerate(axes.flatten()):
    data_hourly = dvar_hourly[dvar_hourly.columns[i]]
    ax.plot(data_hourly, color='blue', linewidth=1)
    # Decorations
    ax.set_title(dvar_hourly.columns[i])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```
